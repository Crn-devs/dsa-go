# String Questions in Go

Reverse A String

```go

func ReverseString(s string) string {

  if len(s) <=0 {
		fmt.Println("Incorrect length please enter a valid string")
  }

  reversed := ""
  for i := len(s)-1; i >= 0; i-- {
    reversed = reversed + string(s[i])
  }
return reversed
}

%%

fmt.Println(ReverseString("Hello World"))

```

Count Vowels in a String

```go
func CountVowels(s string) int {

	if len(s) <= 0 {
		fmt.Println("Incorrect length please enter a valid string between 1-1000 characters")
	}

	Vowels := map[string]int{"a": 1, "e": 2, "i": 3, "o":4, "u":5}

	vCount := 0

	for _, v := range s {
		if _, ok := Vowels[string(v)]; ok {
			vCount++
		}
	}

	return vCount
}


%%
fmt.Println(CountVowels("Hello World")) // 3
fmt.Println(CountVowels("Go Programming")) // 4
```

Check two Strings are equal

```go
import(    "fmt"
    "strings")
func EqualStrings(s, t string) bool {

	if len(s) <= 0 || len(t) <= 0 {
		fmt.Println("incorrect length please enter a valid string")
		return false
	}

	if strings.ToLower(s) == strings.ToLower(t) {
		return true
	}
	return false
}

%%

fmt.Println(EqualStrings("Hello", "hello"))
fmt.Println(EqualStrings("1550, new, old", "1550, new,old"))
```

You are given a string s. Your task is to check if the string is a palindrome. A string is considered a palindrome if it reads the same forward and backward, ignoring spaces, punctuation, and case.
